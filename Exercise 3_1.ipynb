**Purpose:**
Program to read a text file of given sunspot dates and number of spots and plot  
said values on a graph

```python
#Import Libraries
import numpy as np
import matplotlib.pyplot as plt
```

```python
#load text file into a list of lists
solarData = np.loadtxt("C:/Users/slick/School/PH 295/Book Materials/sunspots.txt",float)
```

```python
#Store values in separate lists for graphing and later use
date = solarData[:,0]
spotNum = solarData[:,1]

#Convert dates to int
date = [int(i) for i in date]
```

a) Write program that plots values from sunspots.txt

```python
plt.plot(date,spotNum)
plt.xlabel("time")
plt.ylabel("Sunspots")
plt.title("Sunspots")
plt.show
```

b) Modify your program to only display the first 1000 data points

```python
plt.plot(date,spotNum)
plt.xlim(0,1000)
plt.xlabel("time")
plt.ylabel("Sunspots")
plt.show
```

c) Modify your program further to calc and plot the running average of the data, defined by:

$$ 
Y_k = \frac{1}{2r +1}  \sum_{m=-r}^{r} y_k +m
$$

where r = 5 in this case and y_k are the sunspot numbers.

```python
#initialize list Y
Y = []

#Calculate running average
for i in date:
    r = 5     #reset r
    summ = 0  #reset summ
    for m in range(-r,r+1):
        summ += spotNum[i] + m
    Y.append(1/(2*r+1)*summ)

#Graph and display

#Original Sunspot graph
plt.subplot(2,1,1)
plt.xlim(0,1000)
plt.xlabel("time")
plt.ylabel("Sunspots")
#plt.title("Sunspots over Time")
plt.plot(date,Y)

#Running Average Graph
plt.subplot(2,1,2)
plt.xlim(0,1000)
plt.xlabel("time")
plt.ylabel("Running Average of Sunspots")
#plt.title("Running Average v Time")
plt.plot(date,Y)

#adjust spacing
plt.tight_layout(pad=1.0)
```
