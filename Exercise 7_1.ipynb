**Purpose:**
Calculates the fourier coefficients of given functions using discrete Fourier transforms w/ N = 1000 evenly spaced points then plots their amplitudes.

```python
#Import Libraries
from matplotlib import pyplot as plt
from numpy import linspace, sin, pi, ones_like, zeros, size, fft
from cmath import exp, pi
```

```python
#Define variables
N = 1000
x = linspace(0,1,N)

#Define DFT function
def dft(y):
    c = zeros(N//2+1,complex)
    for k in range(N//2+1):  
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c
```

a) A single cycle of a square wave with amplitude 1

```python
#Calculate y vals for square wave
y1 = ones_like(x)
y1[500:] = -y1[500:]

#DFT of square wave
c1 = dft(y1)

#Plot the figure
plt.figure(dpi = 100)
plt.stem(abs(c1[:20]))
plt.show()
```

b)The sawtooth wave $y_n = n$

```python
#Calculate points for y2
y2 = x

#DFT of Saw Wave
c2 = abs(dft(y2))

plt.figure(dpi = 100)
plt.stem(abs(c2[:20]))
plt.show()
```

c)The modeled sine wave $y_n = \sin(\pi n / N) \sin(20 \pi n/N)$

```python
#Calculate points for y3
y3 = sin(pi*x/N)*sin(20*pi*x/N)

#DFT of sin function
c3 = abs(dft(y3))

#Plot it
plt.figure(dpi = 100)
plt.stem(abs(c3[:20]))
plt.show()
```
