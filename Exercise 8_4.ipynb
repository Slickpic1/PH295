**Purpose:**
Program solves two first-order equations usnig 4th Order Runge Kutta for a pendulum.

*Equations*
$$
\frac{d\theta}{dt} = \omega, \, \, \, \, \frac{d\omega}{dt} = -\frac{g}{l} \sin \theta
$$

```python
#Import Libraries
from numpy import arange, array, sin, pi
from matplotlib import pyplot as plt
```

a) Write a program to solve two first order equations using the fourth-order Runge Kutta method for a pendulum with a 10 cm arm. Use your program to calculate the angle $\theta$ of displacement for several periods of the pendulum when it is released from a standstill at $\theta = 179 degrees$ from vertical. Make a graph of $\theta$ as a function of time.

```python
#Define functions
def f(r,t):
    θ = r[0]
    ω = r[1]
    fθ = ω
    fω = -(g/l)*sin(θ)
    return array([fθ,fω],float)

#RK4
def RK4(r,t):
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    return (k1+2*k2+2*k3+k4)/6

#Define constants
g = 9.81            #m/s^2
l = 0.1             #m

#Define Intervals
a = 0     #start of interval
b = 50    #end of interval
N = 1000
h = (b-a)/N

#Define variables
r = [179*pi/180,0]
ωpoints = []
θpoints = []
tpoints = arange(a,b,h)

#RK4:
for t in tpoints:
    θpoints.append(r[0])
    
    #Do RK4
    r += RK4(r,t)
```

```python
plt.figure(dpi = 100)
plt.plot(tpoints,θpoints)
plt.xlabel("Time (s)")
plt.ylabel("θ (rad)")
plt.title("θ vs Time")
plt.show()
```
