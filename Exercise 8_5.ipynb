**Purpose:**
Program that solves the given equation for $\theta$ as a function of time, and plots it.

*Equation:*
$$
\frac{d^2 \theta}{dt^2} = -\frac{g}{l} \sin \theta + C \cos \theta \sin \Omega t
$$

```python
#Import libraries
from numpy import arange, array, sin, pi, cos
from matplotlib import pyplot as plt
```

```python
#Define functions
def f(r,t):
    θ = r[0]
    ω = r[1]
    fθ = ω
    fω = -(g/l)*sin(θ) + C*cos(θ)*sin(Ω*t)
    return array([fθ,fω],float)

#RK4
def RK4(r,t):
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    
    return (k1+2*k2+2*k3+k4)/6

#Define constants
g = 9.81            #m/s^2
l = 0.1             #m
C = 2               #s^-2
Ω = 5               #s^-1

#Define Intervals
a = 0      #start of interval
b = 100    #end of interval
N = 1000
h = (b-a)/N

#Define variables
r = [0,0]
ωpoints = []
θpoints = []
tpoints = arange(a,b,h)

#RK4:
for t in tpoints:
    θpoints.append(r[0])
    
    #Do RK4
    r += RK4(r,t)
```

```python
#Plot the Figure
plt.figure(dpi = 100)
plt.plot(tpoints,θpoints)
plt.xlabel("Time (s)")
plt.ylabel("θ (rad)")
plt.title("Driven Pendulum")
plt.show()
```
