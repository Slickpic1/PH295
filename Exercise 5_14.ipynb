**Purpose:**
Calculates the gravitational force on a $1 kg$ point mass from a uniform sheet of negligible thickness and mass of $10$ metric tonnes then plots it.

*Gravitational Force:*
$$
F_{z} = G\sigma z \iint_{L/2}^{L/2}\frac{dxdy}{(x^2+y^2+z^2)^{3/2}}
$$
  
*Gauss-Legendre product formula:*
$$
I \simeq \sum_{i=1}^{N} \sum_{j=1}^{N} w_{i} w_{j} f(x_{i},y_{j})
$$

b) Calculate and plot the force as a function of of $z$ from $z = 0$ m to $z = 10$ m. Use above eq with 100 sample points along each axis to get your answer.

```python
#Import Libraries
import matplotlib.pyplot as plt
from gaussxw import gaussxw, gaussxwab
from numpy import linspace
```

```python
#Define constants
G = 6.674e-11
mpoint = 1        #kg
msheet = 1e5      #kg
L = 10
sigma = msheet/(L*L)
```

```python
#define functions
def dxdy(x,y,z):
    return 1/(x**2 + y**2 + z**2)**(3/2)

#force of a given z value
def F(z,dxdy):
    N = 100
    
    #calculate weights and copy
    x,wx = gaussxwab(N,-L/2,L/2)
    y,wy = gaussxwab(N,-L/2,L/2)
    
    #calculate integral and return
    s = 0.0
    for i in range(N):
        for j in range(N):
            s += wx[i]*wy[j]*dxdy(x[i],y[j],z)
    return s*G*sigma*z
```

```python
Force = []
z = linspace(0,10,100)
for k in range(len(z)):
    Force.append(F(z[k],dxdy))
```

```python
plt.plot(z,Force)
plt.title(f'Force on a point particle at height z')
plt.xlabel(f'Z Height(m)')
plt.ylabel(f'Force(N)')
```
