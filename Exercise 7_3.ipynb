**Purpose:** 
Program that loads a waveform, plots it, and calculates it's discrete Fourier Tranform and plots the magnitudes of the first 10000 coefficients using fast fourier tranform (FFT).

*(Possible) Conversion from k-space to f-space:*
$$
\Delta f = \frac{f_s}{N}
$$

```python
#Import Libraries
from numpy import fft, loadtxt, linspace, arange
from matplotlib import pyplot as plt
```

a) Load a waveform file, plot it, calculate its discrete FFT and plot the magnitude of the first 10000 coefficients. Comment on what we can conclude from these graphs about their respective instruments.

```python
#Load Waveform txt file into array
piano = loadtxt('C:/Users/slick/School/PH 295/Book Materials/piano.txt')

#Calculate the DFT coefficients of waveform file
c = fft.fft(piano)

#Graph both the coefficients and the regular waveform file
plt.figure(dpi = 100)
plt.plot(piano)
plt.title(f'Waveform of Piano')
#plt.xlabel(f'time')

plt.figure(dpi = 100)
plt.plot(abs(c[:10000]))
plt.title(f'FFT Coefficients')
plt.xlabel(f'k value')
plt.ylabel(f'FFT Coefficient (ck)')
```

There seems to be a high value of k that corresponds to some frequency at $\sim 1500$ k, followed by some other residual frequencies past 2000 k.

b) From this Fourier Transform, calculate what note was being played on the piano.

```python
#Change from k-space to Frequency-space
fs = 44100       #sampling frequency in Hz
df = fs/len(piano)
f = arange(0,fs,df)

#Plot it again
plt.figure(dpi = 100)
plt.plot(f[1000:1400],abs(c[1000:1400]))
plt.title(f'FFT Coefficients')
plt.xlabel(f'Frequency (Hz)')
plt.ylabel(f'FFT Coefficient (ck)')
```

**Conlusion:**
Frequency seems to match $C_5$ on the note scale. Very weird to see the ability to measure data out of a transform thats something other than gibberish.
