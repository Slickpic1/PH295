*Normal Vector*
$$
v = \nabla \big[z-w(x,y)\big] = \begin {pmatrix} \partial /\partial x \\ \partial /\partial y \\ \partial /\partial z \end {pmatrix} \big[z - w(x,y)\big] = \begin {pmatrix} -\partial w /\partial x \\ -\partial w /\partial y \\ \partial / 1 \end {pmatrix}
$$
  
*Intensity*
$$
I = -\frac{\cos \phi (\partial w /\partial x) + \sin \phi (\partial w /\partial y)}{\sqrt{(\partial w /\partial x)^2 + (\partial w /\partial y)^2 +1}}
$$
  

a) Take the *altitude.txt* file and read it into an array. Then calculate $\frac{\partial{\omega}}{\partial{x}}$ and $\frac{\partial{\omega}}{\partial{y}}$ of $\omega\,(x,y)$ for each gridpoint.  
Explain what method was used to calculate them and why.

```python
#import Libs
from numpy import loadtxt, shape, cos, sin, empty, sqrt,pi
from matplotlib.pyplot import imshow, gray, show,figure
```

```python
#Import txt into array
w = loadtxt("C:/Users/slick/School/PH 295/Book Materials/altitude.txt",float)

#Define Vars
h = 30000          #m
 = 45*pi/180

#find size of array
xsize,ysize = shape(w)
I = empty([xsize,ysize],float)

#calculate partial derivative
for x in range(xsize):
    for y in range(ysize):
        dx = 0
        dy = 0
        
        #forward distance
        if x == 0:
            dx = 0
        elif x == xsize - 1:
            dx = 0
        else:
            dx = (w[x+1,y]-w[x-1,y])/2/h
            
        if y == 0:
            dy = 0
        elif y == ysize - 1:
            dy = 0
        else:
            dy = (w[x,y+1]-w[x,y-1])/2/h
            
        I[x,y] = (cos()*dx + sin()*dy)/sqrt(dx**2+dy**2+1)


imshow(w)
gray()
```

b) Plot the Intensity on a graph

```python
fig = figure(dpi=200)
ax = fig.add_subplot(111)
imshow(I,origin="upper", extent = [0,xsize,0,ysize],cmap = 'gray')
ax.set_aspect(.3)
```
