**Purpose:**
Reads the sunspots.txt file and plots it as a function of time, then calculates the DFT of the data and plots it

```python
#Import Libs
from cmath import exp, pi
from numpy import zeros, loadtxt, linspace, arange
from matplotlib import pyplot as plt
```

```python
#Define DFT Function
def dft(y):
    N = len(y)
    c = zeros(N//2+1,complex)
    for k in range(N//2+1):    #ignor second half coeff because they are not real
        for n in range(N):
            c[k] += y[n]*exp(-2j*pi*k*n/N)
    return c
```

a) Write a program that reads a data in the file and makes a graph of the sunspots

```python
#Import sunspots txt
sunspots = loadtxt('C:/Users/slick/School/PH 295/Book Materials/sunspots.txt',float)

#Plot data
plt.figure(dpi = 100)
plt.plot(sunspots[:,1])
plt.title(f'Sunpots over Time')
plt.xlabel(f'Time of Months')
plt.ylabel(f'Number of Sunspots')
```

b) Calculate the Fourier Transform of sunspots data and make a graph of the magnitude squared $|c_k|^2$ of the Fourier Coefficients as a function of k.

```python
#Calculate coefficients of the Sunspots
c = dft(sunspots[:,1])

#Plot DFT data
plt.figure(dpi=100)
plt.plot(abs(c)**2)
plt.title(f'Power Spectrum')
plt.xlabel(f'k value')
plt.ylabel(f'Magnitude Squared (ck^2)')
plt.show()
```
