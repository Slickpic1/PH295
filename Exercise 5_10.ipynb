**Purpose:**
Calculates the period of an oscillator for a given amplitude using Gaussian quadrature and graphs the period for amplitudes ranging from $a = 0$ to $a = 3$.

*Period:*
$$
T = \sqrt{8m}\int_{0}^{a}\frac{dx}{\sqrt{V(a)-V(x)}}
$$
  
*Potential:*
$$
V(x) = x^4
$$

b) Write a python function that calculates the period of the oscillator for given amplitude $a$ using Gaussian quadrature with $N = 20$ points, then use your function to make a graph of the period for amplitudes ranging from $a = 0$ to $a = 3$.

```python
#Import Libraries
from gaussxw import gaussxw, gaussxwab
import matplotlib.pyplot as plt
from math import sqrt
from numpy import linspace
```

```python
#Define functions
def V(x):
    return x**4

def T(V,b):
    N = 20
    a = 0
    m = 1

    #Calculate the positions, weights, and mapping
    x,w = gaussxwab(N,a,b)
    I = 0.0
    for k in range(N):  
        I += w[k]*(1/sqrt(V(b)-V(x[k])))
    
    I = I*sqrt(8*m)
    return I
```

```python
#Calulate values for the graph with given amplitudes
n = 20
b = linspace(0.1,2,n)
period = []
for i in range(n):
    period.append(T(V,b[i]))
    
plt.plot(b,period,'k.')
plt.title("Period of an anharmonic oscillator")
plt.xlabel("Amplitude (m)")
plt.ylabel("Period (s)")
plt.show
```

*Conclusion:*
As the amplitude gets larger, the period starts to drop significantly and approach 0. As the amplitude reaches 0, the period diverges and approaches infinity. By best estimation, as $a \to 0$, the maximum potential energy $U_{max} \to 0$ as well, driving T higher to conserve energy in the form of kinetic energy. Or maybe not, I could very well be wrong.
