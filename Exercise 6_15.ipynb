**Purpose:**
Find the roots of a degree six polynomial by graphing the function and using those estimates to make an initial guess for a program to calculate with.

*Polynomial*
$$
P(x) = 924x^6 - 2772x^5 +3150x^4 - 1680x^3 + 420x^2 - 42x +1
$$
  
*1st Derivative of Polynomial*
$$
P'(x) = 5544x^5 - 13860x^4 + 12600x^3 - 5040x^2 + 840x - 42
$$

```python
#Import Libraries
from matplotlib import pyplot as plt
from numpy import linspace, arange
```

a) Make a plot of P(x) from $ x=0$ to $x=1$ and by inspecting it find rouch values for the six roots of the polynomial - the points at which the function is 0.

```python
#Define functions
def P(x):
    return 924*x**6-2772*x**5+3150*x**4-1680*x**3+420*x**2-42*x+1
```

```python
#Make list of x values
x = linspace(0,1,100)

#Plot graph
plt.figure(figsize = (13,6))
plt.axhline(y=0, color='r', linestyle='-')
plt.xticks(arange(min(x),max(x),0.05))
plt.xlabel('x')
plt.ylabel('P(x)')
plt.title('Polynomial Root Approximation')
plt.plot(x,P(x))
```

b) Write a python program to solve for the positions of all six roots to at least ten decimal places of accuracy using Newtons Method.

```python
#Define first derivative of P(x)
def dPdx(x):
    return 5544*x**5 - 13860*x**4 + 12600*x**3 - 5040*x**2 + 840*x - 42

#Define parameters
accuracy = 1e-10
xguess = [0.05,0.16,0.37,0.625,0.83,0.97]    #guessed values of approx roots

for i in range(len(xguess)):
    delta = 1.0     #reset delta for each iteration
    xroot = xguess[i]
    while abs(delta)>accuracy:
        delta = P(xroot)/dPdx(xroot)
        xroot -= delta
    print(f'Root {i+1} of P(x) is {xroot}')
```

**Conclusion**
Newtons method seems to be a great approximation for how to find the roots of functions, so long as we can find the derivative of the functions. Only problems are that the answer is not exact, but an approximate of the root.
