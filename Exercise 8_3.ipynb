**Purpose:**
Program that solves the Lorentz equations below with given constants across a specific range, then plot them.

*Lorenz Equations*
$$
\frac{dx}{dt} = \sigma (y-x), \, \, \, \frac{dy}{dt} = cx - y - xz, \, \, \, \frac{dz}{dt} = xy - bz
$$

```python
#Import Libraries
from numpy import arange, array
from matplotlib import pyplot as plt
```

a) Write a prgram to solve the Lorenz Transform for the case $\sigma = 10$, $c = 28$, and $b = \frac{8}{3}$ from $t = 0$ to $t = 50$ with initial conditions $(x,y,z) = (0,1,0).$

```python
#Define functions
def f(r,t):
    x = r[0]
    y = r[1]
    z = r[2]
    fx = σ*(y-x)
    fy = c*x-y-x*z
    fz = x*y - β*z
    return array([fx,fy,fz])

#RK4
def RK4(r,t):
    k1 = h*f(r,t)
    k2 = h*f(r+0.5*k1,t+0.5*h)
    k3 = h*f(r+0.5*k2,t+0.5*h)
    k4 = h*f(r+k3,t+h)
    return (k1+2*k2+2*k3+k4)/6

#Define constants
σ = 10
c = 28
β = 8.0/3.0

#Define intervals
a = 0       #start interval
b = 50      #end interval
N = 1000    #Number of steps
h = (b-a)/N # Steps size

#Define initial conditions
r = [0,1,0]    #starting x,y,z

#Define lists
tpoints = arange(a,b,h)
xpoints = []
ypoints = []
zpoints = []

#RK4
for t in tpoints:
    xpoints.append(r[0])
    ypoints.append(r[1])
    zpoints.append(r[2])
    
    #Do Runge Kutta
    r += RK4(r,t)
```

```python
#Plot y as a function of time
plt.figure(dpi = 100)
plt.plot(tpoints,ypoints)
plt.xlabel("Time (s)")
plt.ylabel("y Position")
plt.title("y Position vs Time")
plt.show()
```

b) Modify your program to produce a plot of z against x.

```python
#Plot z against x
plt.figure(dpi = 100)
plt.plot(xpoints,zpoints)
plt.xlabel("x Position")
plt.ylabel("z Position")
plt.title("Strange Attractor")
plt.show()
```
