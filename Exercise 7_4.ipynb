**Purpose:**
Reads and converts data from dow.txt to Fourier Coefficients and filters and smooths the values.

```python
#Import Libraries
from numpy import linspace, loadtxt, arange, fft, zeros
from matplotlib import pyplot as plt
```

a) Read in the data from dow.txt and plot it on a graph

```python
#Import Data from dow.txt
dow = loadtxt('C:/Users/slick/School/PH 295/Book Materials/dow.txt')

#Plot it
plt.figure(dpi = 100)
plt.plot(dow)
plt.title(f'Dow Jones Industrial Average')
plt.xlabel(f'Time (days)')
plt.ylabel(f'Closing Value (Human Souls)')
```

b) Calculate the coefficients of the discrete Fourier Transform of the data. Note that it produces an array of $\frac{1}{2}N + 1$ complex numbers.

```python
#Calculate Fourier Coefficients
c = fft.fft(dow)
```

c) Now set all but the first 10% of this array to zero.

```python
#Find length of c, calculate 10% of that value, round down, and set remaining to zero
cutoff = int(len(c)*0.10)
c[cutoff:] = 0
```

d) Calculate the inverse Fourier Transform of the resulting array and plot it on the same graph as the original.

```python
#Calculate IFFt of c
newDow = fft.irfft(c,len(c))

#Plot old and new graph
plt.figure(dpi = 100)
plt.plot(dow)
plt.plot(newDow)
plt.title(f'Dow Jones vs. Smoothed Dow Jones')
plt.xlabel(f'Time (days)')
plt.ylabel(f'Daily Closing Value (in teeth)')
```

e) Modify your program so that it sets all but the first 2% of the coefficients and run it again.

```python
#Calculate Fourier Coefficients
c = fft.fft(dow)

#Find length of c, calculate 10% of that value, round down, and set remaining to zero
cutoff = int(len(c)*0.02)
c[cutoff:] = 0

#Calculate IFFt of c
newDow = fft.irfft(c,len(c))   #Required to fit them ontop of one another, otherwise returns twice the ammount (stretched)

#Plot old and new graph
plt.figure(dpi = 100)
plt.plot(dow)
plt.plot(newDow)
plt.title(f'Dow Jones vs. Smoothed Dow Jones')
plt.xlabel(f'Time (days)')
plt.ylabel(f'Daily Closing Value (in teeth)')
```
