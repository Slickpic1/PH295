**Purpose:**
Uses combined overrelatation/Gauss-Seidel method to solve Laplace's equation for the 2d problem in Example 9.1.

*Gauss Seidel and Overrelaxation*
$$
\phi(x,y) \leftarrow \frac{1+ \omega}{4}\big[\phi(x+a,y) + \phi(x-a,y) + \phi(x,y+a) + \phi(x,y-a)\big] - \omega \phi(x,y)
$$
  
*Laplace Equation*
$$
\frac{\partial^2 \phi}{\partial x^2} + \frac{\partial^2 \phi}{\partial y^2} = 0 
$$

```python
#Import Libraries
from numpy import arange, zeros
from matplotlib import pyplot as plt
```

```python
#Define Constants
L = 1              #length in m
M = 100         
V = 1              #Voltage of the side
ω = 0.9            #Parameter we are changing ϕ by
a = 0.01
target = 1e-6

#Build Arrays
ϕ = zeros([M+1,M+1],float)
ϕ[0,:] = V                  #boundary condition for the top
```

```python
#Define error
error = 1.0

#Gauss-Seidel Method
while error > target:
    error = 0.0   #reset error
    for i in range(1,M):
        for j in range(1,M):
            ϕold = ϕ[i,j]   #do i need to reset?
            ϕ[i,j] = (1+ω) * (ϕ[i+1,j]+ϕ[i-1,j]+ϕ[i,j+1]+ϕ[i,j-1])/4 - ω*ϕ[i,j] 
    
        #Calculate the error of the step
            error = max(abs(ϕold - ϕ[i,j]),error)
```

```python
#Plot the stuff
plt.imshow(ϕ)
plt.gray()
plt.colorbar()
plt.show()
```
